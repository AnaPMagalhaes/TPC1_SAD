```
def solution(s):
  n = []
  for c in s:
    if c == ')':
      if n[-1] == '(':
        n.pop()
      else:
        return 0
    elif c == '}':
      if n[-1] == '{':
        n.pop()
      else:
        return 0
    elif c == ']':
      if n[-1] == '[':
        n.pop()
    else:
      n.append(c)
  return 1
print (solution ('[({})]'))
```
